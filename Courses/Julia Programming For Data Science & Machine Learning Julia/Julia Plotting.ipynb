```julia
year = [x for x in range(2000, length= 10)]
```

```julia
sale = 100 * rand(10)
```

```julia
plot(year, sale, labels="Line Plot")
scatter!(year, sale, labels="Scatter Plot")
xlabel!("Year")
ylabel!("Sale")
title!("Sale between 2000 and 2009")
```

```julia
x = -5:0.2:5
y = x.^2
plot(x,y)
```

```julia
p1 = plot(x, x)
p2 = plot(y, x)
p3 = plot(sin, x)
p4 = plot(cos, x)
plot(p1, p2, p3, p4, layout =(1,4))
```

```julia
plot(sin, x)
plot!(cos, x)
```
